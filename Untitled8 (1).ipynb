```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sales_by_day = df.groupby('day')['total_bill'].sum()
print(sales_by_day)
```

```python
import matplotlib.pyplot as plt

plt.figure()
sales_by_day.plot(kind='bar')

plt.title('Total Sales by Day')
plt.xlabel('Day')
plt.ylabel('Total Sales')

plt.show()
```

```python
top_day = sales_by_day.idxmax()
top_value = sales_by_day.max()

print("Top sales day is:", top_day)
print("Total sales:", top_value)
```
